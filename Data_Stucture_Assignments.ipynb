# Product Pricing

```python
#1
price_dict = {"mattress" : 30000, "pillow": 6000, "mobile": 159000, "watch": 4500, "stove": 3000, "grinder": 4000, "table": 2000, "chair": 1500}
print(price_dict)
```

```python
#2
price_dict["t-shirt"] = 600
price_dict["mattress"] = 27000
del price_dict["chair"]
print(price_dict)
```

```python
#3
average = sum(price_dict.values()) / len(price_dict)
print(average)
```

```python
#optional
minimum = min(price_dict.values())
maximum = max(price_dict.values())
print(minimum)
print(maximum)
```

# combined operations

```python
#1
catalog = [
    (products[0], price_dict[products[0]], "bed room products"),
    (products[1], price_dict[products[1]], "bed room products"),
    (products[2], price_dict[products[2]], "electronics"),
    (products[3], price_dict[products[3]], "electronics"),
    (products[4], price_dict[products[4]], "kitchen products"),
    (products[5], price_dict[products[5]], "kitchen products")
]
print(catalog)
```

```python
category_to_products = {
    "bed room products": ["mattress", "pillow"],
    "electronics": ["mobile", "watch"],
    "kitchen products": ["stove", "grinder"]
}
print(category_to_products)
```
