# Process multiple orders

```python
orders = [1200, 2500, 800, 1750, 3000]  

for i in orders:
    if i >= 2000:
        discount_final = i - i * 0.15
        print(f"order amount is :{i}, with 15% discount final amount is:{discount_final}")

    elif i < 2000 and i >= 1500:
        discount_final = i - i * 0.1
        print(f"order amount is :{i}, with 10% discount final amount is:{discount_final}")

    
    elif i < 1500 and i >= 1000:
        discount_final = i - i * 0.07
        print(f"order amount is :{i}, with 7% discount final amount is:{discount_final}")
    
    else:
        discount_final = i
        print(f"order value is {discount_final}")
```

```python
orders = [1200, 2500, 800, 1750, 3000]  

total_revenue = 0

for i in orders:

    if i >= 2000:
        discount_final = i - i * 0.15

    elif i < 2000 and i >= 1500:
        discount_final = i - i * 0.1
    
    elif i < 1500 and i >= 1000:
        discount_final = i - i * 0.07
    
    else:
        discount_final = i

    total_revenue = total_revenue + discount_final


print(f"sum of orders after discount: {total_revenue}" )
```

# User Menu

```python
order_amount = []

print(f'''
      USER MENU
      1. Add order amount
      2. Show all orders and total after applying discount
      q. Quit
      '''  )

while True:
    choise = input("Kindly choose the above any: ")


    if choise == "1":
        amount = int(input("Enter the order amount: "))
        order_amount.append(amount)
        

    elif choise == "2":
        total_amount = 0

        for i in order_amount:
            if i >= 2000:
                discount_final = i - i * 0.15

            elif i < 2000 and i >= 1500:
                discount_final = i - i * 0.1
            
            elif i < 1500 and i >= 1000:
                discount_final = i - i * 0.07
            
            else:
                discount_final = i

            total_amount = total_amount + discount_final
        print(f"the list of orders are: {order_amount} and the total amount after discount is: {total_amount}")

    elif choise == "q":
        print("User Menu exit successfully!!")
        break

    else: 
        print("Entered choise is invalid, please select from the above again!!")
```

# Loop Control with conditions

```python
daily = [200, 150, 0, 400, 50, -1, 300]

amount_sales = 0

for i in daily:

    if i < 0:
        print(f"It was a currupted data: {i}")
        break

    elif i == 0:
        print(f"Day with no sales: {i}")
        continue

    elif i > 0:      
        amount_sales = amount_sales + i
        
print(f" the total sales are: {amount_sales}")
```

```python
daily = [200, 150, 0, 400, 50, -1, 300]

amount_sales = 0

for i in daily:

    if i < 0:
        break

    elif i == 0:
        continue

    elif i > 0:      
        amount_sales = amount_sales + i
        
print(f" the total sales are: {amount_sales}")
```
